```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
producer = KafkaProducer(bootstrap_servers=['13.200.253.171:9092'],

                         value_serializer=lambda x:
                         dumps(x).encode('utf-8'))
```

```python
df=pd.read_csv("/content/drive/MyDrive/BD/data.csv")
```

```python
df.head()
```

```python
for index, row in df.iterrows():
    # Convert row to dictionary
    row_dict = row.to_dict()
    producer.send('demo_testing2', value=row_dict)
    print(index)
    sleep(1)
```

```python
producer.flush()
```
